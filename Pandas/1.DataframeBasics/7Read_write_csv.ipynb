```python
import pandas as pd
```

```python
df = pd.read_csv('Book1.csv')
df
```

```python
#|---------------------------------------------------------------------------|#
#| IF IT CONTAINS EXTRA ROWS ABOVE
#|---------------------------------------------------------------------------|#
df = pd.read_csv('Book1.csv',skiprows=2)
df
```

```python
#|---------------------------------------------------------------------------|#
#| IF THERE ARE NO HEADERS (column names)
#|---------------------------------------------------------------------------|#
df = pd.read_csv('Book1.csv',header=None,names=['specify column names'])
df
```

```python
#|---------------------------------------------------------------------------|#
#| TO CONVERT NOT AVAILABLE ROWS TO NaN
#|---------------------------------------------------------------------------|#
df = pd.read_csv('Book1.csv',skiprows=2,na_values={
  'eps':['not available','n.a.'],
  'revenue':['not available','n.a.',-1],
  'price':['not available','n.a.'],
  'people':['not available','n.a.']
})
df
```

```python
#|---------------------------------------------------------------------------|#
#| WRITE TO CSV
#|---------------------------------------------------------------------------|#

df.to_csv('new.csv',index=False)
```
